2. What is an interpreter in Python?
   - A Python interpreter is a computer program that executes Python statements one at a time compared to a compiler, which translates the entire code in one go. A Python interpreter translates Python code into machine code that the computer can easily comprehend and execute.
   How the Python Interpreter Operates: Interprets the Python source code (in .py files). Translates and converts the code into an intermediate form called bytecode. Executes the bytecode in the Python Virtual Machine (PVM), which interacts with the computer operating system. General Python Interpreters CPython – The default and most used Python interpreter. PyPy – JIT compiled interpreter with improved speed. Jython – Runs Python programs inside the Java Virtual Machine (JVM). IronPython – Interoperates with.NET and executes Python on the.NET platform. MicroPython – For microcontrollers and embedded systems. You can run the Python interpreter in interactive mode by simply typing python (or python3) at your terminal or command line.

7. What is the difference between “==” and “is” operators in Python?
  - In Python, == and is are both comparison operators, but they serve  
     different purposes:

     1- == (Equality Operator)

     Checks if two values are equal. Compares the contents (or values) of two objects. Example a = [1, 2, 3] b = [1, 2, 3] print(a == b) # True (because their contents are the same)

     2- (Identity Operator)

     Checks if two variables refer to the same object in memory. Compares the identity (memory address) of two objects. Example a = [1, 2, 3] b = [1, 2, 3] print(a is b) # False (because they are different objects in memory)

10. What is the difference between implicit and explicit type casting?
   - 1-> Implicit Type Casting (Type Promotion) Done automatically by the compiler. Happens when converting a smaller data type to a larger one. No data loss occurs. Example: int a = 10; float b = a; // Implicit conversion from int to float.

     2 -> Explicit Type Casting (Type Conversion) Done manually by the programmer. Used when converting a larger data type to a smaller one. May cause data loss or precision loss. Example: float x = 10.99; int y = (int)x; // Explicit conversion from float to int (truncates decimal part).

12. How does the elif statement work?
    -  Python elif is "else if." It is used when you have more than one condition to check and you would prefer to execute various parts of code depending on which condition is True.

     How it works: The program then checks the if condition. If the if condition is True, then that piece of code runs and the rest of the elif and else statements are skipped. When the if condition is False, then the code jumps to the elif condition. If elif condition is True, its block will be executed and others will be skipped. 13. What is the difference between for and while loops?
   -  The main difference between a for loop and a while loop is that a for loop is used when you know the number of iterations beforehand, while a while loop is used when the number of iterations is unknown and depends on a condition. Here's a more detailed explanation: For Loop: Used when you know exactly how many times you want the code to repeat. Typically used to iterate over a sequence (like a list or a range of numbers). The loop's structure includes initialization, condition, and increment/decrement statements in a single line, making it concise and easy to read. Example: for i in range(5): print(i) (This will print numbers 0 to 4) While Loop: Used when you don't know how many times the code needs to repeat, and the loop continues until a specific condition is met. The loop's condition is checked at the beginning of each iteration, and the loop continues as long as the condition is true. Example: count = 0; while count < 5: print(count); count += 1 (This will print numbers 0 to 4).

14. Describe a scenario where a while loop is more suitable than a for loop?
   - A while loop is more suitable than a for loop when the number of iterations is not known in advance and the loop should continue running based on a condition rather than a fixed range.

     Scenario: User Input Validation Imagine you're writing a program that asks a user to enter a positive number. If the user enters a negative number or zero, the program should keep asking untill they enter a valid positive number.

```python
#4. Write a program that checks if a given word is a Python keyword.
import keyword

def is_keyword(word):
  """Checks if a word is a Python keyword.

  Args:
    word: The word to check.

  Returns:
    True if the word is a keyword, False otherwise.
  """
  return keyword.iskeyword(word)

# Example usage
word1 = "for"
word2 = "example"

print(f"'{word1}' is a keyword: {is_keyword(word1)}")
print(f"'{word2}' is a keyword: {is_keyword(word2)}")
```

```python
#10. Write code to demonstrate type casting with list elements.
# Initial list with mixed data types
my_list = ['1', '2', '3.14', 4, 5.6]

# Type casting elements to integers
int_list = [int(x) for x in my_list if isinstance(x, str) and x.isdigit()]
# Type casting elements to floats
float_list = [float(x) for x in my_list if isinstance(x, str)]
# Type casting elements to strings
str_list = [str(x) for x in my_list]

# Printing the results
print("Original List:", my_list)
print("Integer List:", int_list)
print("Float List:", float_list)
print("String List:", str_list)
```

```python
#14. Write a program to reverse a string using a while loop.
def reverse_string(s):
    reversed_s = ""
    length = len(s) - 1
    while length >= 0:
        reversed_s += s[length]
        length -= 1
    return reversed_s

input_string = "Hello, pwskills!"
reversed_string = reverse_string(input_string)
print(f"Original string: {input_string}")
print(f"Reversed string: {reversed_string}")
```

```python
#15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
def factorial(num):
    """
    Calculates the factorial of a given number using a while loop.
    """
    if num < 0:
        return "Factorial is not defined for negative numbers"
    result = 1
    i = 1
    while i <= num:
        result *= i
        i += 1
    return result

# Get user input
number = int(input("Enter a number to find its factorial: "))

# Calculate and print the factorial
print(f"The factorial of {number} is: {factorial(number)}")
```
